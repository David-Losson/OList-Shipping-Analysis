# Create Scatterplots

### Relationship between Delivery Time and Delivery to the Carrier Time

```python
#Create a scatterplot for the Purchase-Delivery and Estimated-Purchase Time Differences

sns.lmplot(x = 'Delivery-Purchase_Time_Difference_Min', y = 'Carrier_Delivered-Order_Time_Difference_Min', data = df_5)
plt.show()
```

Excepting the few outliers, the trendline indicates to us that as the the time lag between ordering and delivery to the carrier increases, overall fulfillment time increases, up to a certain point. Beyond 87 days or so (i.e. 125,000 minutes), the shipments are not affected by the lag in delivery to the carrier as much, rather, other factors in the shipping process may account for the prolonged time.

### Relationship Between Delivery Time and Estimated Delivery Time by Review Rating

```python
#Create a scatterplot for the Purchase-Delivery Time Difference and Review Score

sns.lmplot(x = 'Delivery-Purchase_Time_Difference_Min', y = 'Review_Score', data = df_5)
plt.show()
```

While there is a distinct concentration of longer fulfillment times at the lowest possible review score (1), the relationship between fulfillment time and review score is less clear, as the scores increase (i.e. 2-5.)

### Relationship between Price and Freight Value

```python
#Create a scatterplot for the Price and Freight Value

sns.lmplot(x = 'Price', y = 'Freight_Value', data = df_5)
plt.show()
```

Unsurprisingly, there is a clear, direct relationship between price and the value of the freight shipping cost.

### Relationship between Freight Value and Product Weight

```python
#Create a scatterplot for Freight Value and Product Weight

sns.lmplot(x = 'Freight_Value', y = 'Product_Weight(g)', data = df_5)
plt.show()
```

There are basically no products beyond 30,000 grams (66 lbs) in weight. Generally, the heavier the product, the higher freight value but that's only somewhat true. Plenty of items of different weights have the same freight value.

### Relationship between Estimated Fulfillment Time and Actual Fulfillment Time

```python
#Create a scatterplot for the Estimated Delivery-Purchase Time Difference and the Delivery-Purchase Time Difference

sns.lmplot(x = 'Delivery-Purchase_Time_Difference_Min', y = 'Estimated-Purchase_Time_Difference_Min', data = df_5)
plt.show()
```

The estimated delivery time becomes more accurate with respect to the actual purchase-delivery time, as you get closer to the 50,000 minute-mark (35 days). There is a large number of orders with estimated delivery times under 50,000 minutes (35 days) for which the actual delivery time exceeds the estimated delivery time.
