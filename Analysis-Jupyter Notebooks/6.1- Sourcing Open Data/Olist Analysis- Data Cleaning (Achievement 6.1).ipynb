```python
#Convert the delivered, order purchase, order approved, estimated delivery, and order delivered to carrier dates to datetime format
df_geolocation_customers_sellers_items_orders_clean['order_purchase_timestamp'] = df_geolocation_customers_sellers_items_orders_clean['order_purchase_timestamp'].astype('datetime64[ns]')
df_geolocation_customers_sellers_items_orders_clean['order_delivered_customer_date'] = df_geolocation_customers_sellers_items_orders_clean['order_delivered_customer_date'].astype('datetime64[ns]')
df_geolocation_customers_sellers_items_orders_clean['order_estimated_delivery_date'] = df_geolocation_customers_sellers_items_orders_clean['order_estimated_delivery_date'].astype('datetime64[ns]')
df_geolocation_customers_sellers_items_orders_clean['order_delivered_carrier_date'] = df_geolocation_customers_sellers_items_orders_clean['order_delivered_carrier_date'].astype('datetime64[ns]')
df_geolocation_customers_sellers_items_orders_clean['order_approved_at'] = df_geolocation_customers_sellers_items_orders_clean['order_approved_at'].astype('datetime64[ns]')
```

```python
df_geolocation_customers_sellers_items_orders_clean.info()
```

```python
df_geolocation_customers_sellers_items_orders_clean
```

```python
#Calculate the time difference between actual delivery date and order purchase date
df_geolocation_customers_sellers_items_orders_clean ['Delivery-Purchase_Time_Difference'] = df_geolocation_customers_sellers_items_orders_clean['order_delivered_customer_date'] - df_geolocation_customers_sellers_items_orders_clean['order_purchase_timestamp']

#Calculate the time difference between estimated delivery date and order purchase date
df_geolocation_customers_sellers_items_orders_clean ['Estimated-Purchase_Time_Difference'] = df_geolocation_customers_sellers_items_orders_clean['order_estimated_delivery_date'] - df_geolocation_customers_sellers_items_orders_clean['order_purchase_timestamp']

#Calculate the time difference between order approval date and order purchase date
df_geolocation_customers_sellers_items_orders_clean ['Approval-Purchase_Time_Difference'] = df_geolocation_customers_sellers_items_orders_clean['order_approved_at'] - df_geolocation_customers_sellers_items_orders_clean['order_purchase_timestamp']

#Calculate the time difference between delivery to the carrier and order purchase date
df_geolocation_customers_sellers_items_orders_clean ['Carrier_Delivered-Order_Time_Difference'] = df_geolocation_customers_sellers_items_orders_clean['order_delivered_carrier_date'] - df_geolocation_customers_sellers_items_orders_clean['order_purchase_timestamp']
```

```python
#Convert the calculated time differences into minute amounts in the integer or float format
df_geolocation_customers_sellers_items_orders_clean ['Delivery-Purchase_Time_Difference_Min'] = df_geolocation_customers_sellers_items_orders_clean ['Delivery-Purchase_Time_Difference']/pd.Timedelta(minutes=1)
df_geolocation_customers_sellers_items_orders_clean ['Estimated-Purchase_Time_Difference_Min'] = df_geolocation_customers_sellers_items_orders_clean ['Estimated-Purchase_Time_Difference']/pd.Timedelta(minutes=1)
df_geolocation_customers_sellers_items_orders_clean ['Approval-Purchase_Time_Difference_Min'] = df_geolocation_customers_sellers_items_orders_clean ['Approval-Purchase_Time_Difference']/pd.Timedelta(minutes=1)
df_geolocation_customers_sellers_items_orders_clean ['Carrier_Delivered-Order_Time_Difference_Min'] = df_geolocation_customers_sellers_items_orders_clean ['Carrier_Delivered-Order_Time_Difference']/pd.Timedelta(minutes=1)
```
